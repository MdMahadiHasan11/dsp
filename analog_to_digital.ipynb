```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 4)
```

```python
A = 10
f = 100
T = 1/f
t = np.arange(0, 2*T + T/100, T/100)
y = A*np.sin(2*np.pi*f*t)

plt.plot(t, y)
plt.yticks(np.arange(-A, A + 1, 2))
plt.title('Transmitting Signal')
plt.xlabel('Time (sec)')
plt.ylabel('Ampitude (volt)')
plt.show()
```

```python
Ts = T/20
Fs = 1/Ts
n = np.arange(0, T/Ts + 1, 1)
y = A*np.sin(2*np.pi*f*n*Ts)

plt.stem(n, y)
plt.xticks(n)
plt.yticks(np.arange(-A, A + 1, 2))
plt.title('Discrete Time Signal After Sampling')
plt.xlabel('Time (sec)')
plt.ylabel('Ampitude (volt)')
plt.show()
```

```python
y_new = y + A
plt.stem(n, y_new, markerfmt= 'o')
plt.xticks(n)
plt.yticks(np.arange(0, 2*A + 1, 2))
plt.title('Dc Level + Discrete Time Signal After Sampling')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (volt)')
plt.show()
```

```python
y_ = np.round(y + A)
plt.stem(n, y_, markerfmt= 'o')
plt.yticks(np.arange(0, 2*A + 1, 2))
plt.title('Quantization of Signal')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (volt)')
plt.show()
```
